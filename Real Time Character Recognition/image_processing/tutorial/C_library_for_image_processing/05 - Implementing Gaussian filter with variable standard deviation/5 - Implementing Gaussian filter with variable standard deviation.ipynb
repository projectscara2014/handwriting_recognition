# how the gaussian filter looks

```python
def time_this_function(function) : 
    try : 
        time
    except : 
        import time
    
    def wrapper_function(*args,**kwargs) : 
        start_time = time.time()
        return_value = function(*args,**kwargs)
        elapsed_time = (time.time() - start_time)*1000
        print(function.__name__ + ' -- elapsed_time -- ' + str(elapsed_time),end=' ms\n')
        return return_value
    return wrapper_function

@time_this_function
def c_test(image,kernel,kn) : 
    c_image = libconvolution.convolution(image,k,kn)
    return c_image
```

```python
import gaussian
import libconvolution
import image_processing

image = image_processing.read_image('sample_image_3.png')
image_processing.display(image)
g = gaussian.gaussian()
```

```python
k = g.generate_gaussian_smoothening_kernel(1)
kn = sum([sum(element) for element in k])
print('for sigma = ',0.1)
image_processing.display(numpy.array(k))
```

```python
image_processing.display(c_test(image,k,kn))
```

```python
k = g.generate_gaussian_smoothening_kernel(5)
print('for sigma = ',5)
plt.imshow(numpy.array(k),cmap=matplotlib.pyplot.get_cmap('gray'))
kn = sum([sum(element) for element in k])
```

```python
image_processing.display(c_test(image,k,kn))
```
